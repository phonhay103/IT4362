```python
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv(
    'goodreads-300k-dataset/goodreads.csv',
    index_col=0,
)
```

```python
# Pandas-profiling
# from pandas_profiling import ProfileReport
# profile = ProfileReport(df, title="Pandas Profiling Report")
# profile.to_file("pandas_profiling_report.html")
```

```python
sns.displot(df.rating, kde=True, bins=150).savefig("rating_K0.png")
```

```python
df1 = df[df.rating_count >= 1000]
sns.displot(df1.rating, kde=True, bins=150).savefig("rating_K1000.png")
```

```python
bins = (0, 3, 3.5, 4, 4.5, 5)
labels = np.arange(len(bins)-1)
labels = ['(0, 3.0]', '(3.0, 3.5]', '(3.5, 4.0]', '(4.0, 4.5]', '(4.5, 5.0]']
sns.displot(pd.cut(df1.rating, bins=bins, labels=labels, include_lowest=True)).savefig("rating_K1000_category.png")
```
