```python
import pandas as pd
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv(
    'goodreads-300k-dataset/keywords.csv', 
    usecols = ['rating', 'rating_count']
)
```

```python
sns.displot(df.rating, kde=True)
```

```python
df1 = df[df.rating_count >= 100]
sns.displot(df1.rating, kde=True)
```

```python
df2 = df[df.rating_count >= 1000]
sns.displot(df2.rating, kde=True)
```

```python
bins = (0, 3, 3.5, 4, 4.5, 5)
labels = np.arange(len(bins)-1)
sns.displot(pd.cut(df.rating, bins=bins, labels=labels, include_lowest=True))
```

```python
bins = (0, 3, 3.5, 4, 4.5, 5)
labels = np.arange(len(bins)-1)
sns.displot(pd.cut(df1.rating, bins=bins, labels=labels, include_lowest=True))
```

```python
bins = (0, 3, 3.5, 4, 4.5, 5)
labels = np.arange(len(bins)-1)
sns.displot(pd.cut(df2.rating, bins=bins, labels=labels, include_lowest=True))
```
